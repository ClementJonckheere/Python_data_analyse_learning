# **LES FILTRES**

```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df["gender"] == "Male"
```

## **Filtre les rangés qui non pas la valeur Male**

```python
df[df["gender"] == "Male"]
```

## **Filtre des pays**

```python
country_filter = df["country"].isin(["France","Canada"])
df[country_filter]
```

## **Filtre des prix**

```python
df_paid = df.copy()
```

```python
df_paid.price_paid = df_paid.price_paid.apply(lambda x: x.replace("$",""))
df_paid.price_paid = df_paid.price_paid.astype(float)
```

```python
df_paid[df_paid["price_paid"] >= 5.0]
```
