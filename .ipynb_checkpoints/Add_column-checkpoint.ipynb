```python
import pandas as pd
df = pd.read_csv("data.csv")
```

## **Remplacer les valeurs null par 0**

```python
df["tax"].fillna(0, inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
df.price_paid = df.price_paid.apply(lambda x: x.replace("$",""))
df.price_paid = df.price_paid.astype(float)
```

```python
df
```

```python
df.dtypes
```

## **Rajouter une column**

```python
df['test'] = 0
```

```python
df
```

```python
del df["test"]
```

## **Ajout d'une column et ajouter la taxe au prix**

```python
df["price_total"] = df["price_paid"] * (1 - df["tax"] /100)
```

```python
df
```

## **Manipulation pays pour ajouter des codes pays**

```python
countries = {"United States":"UN", "France":"FR", "Canada":"CA", "Marocco":"MA"}
df['country_code'] = df['country'].map(countries)
```
